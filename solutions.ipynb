## Set up environment

```python
import cv2
import matplotlib
import numpy as np

img_path = "room.jpg"

def get_img():
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = get_img()
matplotlib.pyplot.imshow(img)
```

## Task 1
Convert the image into grayscale

```python
img = get_img()
grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
matplotlib.pyplot.imshow(grayscaled, cmap='gray')
```

## Task 2
Convert the image color space into hsv

```python
img = get_img()
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
matplotlib.pyplot.imshow(hsv)
```

## Task 3
Flip the image vertically

```python
img = get_img()
flipped = cv2.flip(img, 0)
matplotlib.pyplot.imshow(flipped)
```

## Task 4
Flip the image horizontally

```python
img = get_img()
flipped = cv2.flip(img, 1)
matplotlib.pyplot.imshow(flipped)
```

## Task 5
Rotate the image by 45

```python
img = get_img()
angle = 45
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
scale = 1.0
matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, matrix, (w, h))
matplotlib.pyplot.imshow(rotated)
```

## Task 6
Rotate the image by 30 around a specific point

```python
img = get_img()
angle = 30
(h, w) = img.shape[:2]
point = (w // 4, h // 4)
scale = 1.0
matrix = cv2.getRotationMatrix2D(point, angle, scale)
rotated = cv2.warpAffine(img, matrix, (w, h))
matplotlib.pyplot.imshow(rotated)
```

## Task 7
Shift the image right by 10 pixels

```python
img = get_img()
shift_x = 10
shift_y = 0
transition_matrix = np.float32([
	[1, 0, shift_x],
	[0, 1, shift_y]
])
(h, w) = img.shape[:2]
shifted = cv2.warpAffine(img, transition_matrix, (w, h))
matplotlib.pyplot.imshow(shifted)
```

## Task 8
Change brightness of the image

```python
img = get_img()
gain = 1
bias = 120
brightned = cv2.convertScaleAbs(img, alpha=gain, beta=bias)
matplotlib.pyplot.imshow(brightned)
```

## Task 9
Change contrast of the image

```python
img = get_img()
gain = 2.5
bias = 0
contrasted = cv2.convertScaleAbs(img, alpha=gain, beta=bias)
matplotlib.pyplot.imshow(contrasted)
```

## Task 10
Binarize the image

```python
img = get_img()
grayscaled = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
thresh = 127
max_val = 255
_, binary = cv2.threshold(grayscaled, thresh, max_val, cv2.THRESH_BINARY)
matplotlib.pyplot.imshow(binary,  cmap='gray')
```

## Task 11
Extract SIFT keypoints from the image

```python
img_with_keypoints = get_img()
grayscaled = cv2.cvtColor(img_with_keypoints, cv2.COLOR_RGB2GRAY)
sift = cv2.SIFT_create()
keypoints = sift.detect(grayscaled, None)
cv2.drawKeypoints(grayscaled, keypoints, img_with_keypoints)
matplotlib.pyplot.imshow(img_with_keypoints)
```

## Task 12
Extract ORB keypoints from the image

```python
img_with_keypoints = get_img()
grayscaled = cv2.cvtColor(img_with_keypoints, cv2.COLOR_RGB2GRAY)
orb = cv2.ORB_create()
keypoints = orb.detect(grayscaled, None)
cv2.drawKeypoints(grayscaled, keypoints, img_with_keypoints)
matplotlib.pyplot.imshow(img_with_keypoints)
```
